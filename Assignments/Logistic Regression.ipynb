```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
```

```python
df.rename(columns={df.columns[-1]: 'Annual Income'}, inplace=True)
```

```python
df
```

```python
# Step 3: Change categorical data to numerical data using Label Encoder
label_encoders = {}
categorical_columns = df.select_dtypes(include=['object']).columns

for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
label_encoders = {}
categorical_columns = df.select_dtypes(include=['object']).columns

for col in categorical_columns:
    le = LabelEncoder()
    df.loc[:, col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
df
```

```python
X = df.drop('Annual Income', axis=1)
y = df['Annual Income']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train, y_train)
```

```python
accuracy = logistic_model.score(X_test, y_test)
print(f'Logistic Regression Model Accuracy: {accuracy:.2f}')
```

```python
y_pred = logistic_model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
tn, fp, fn, tp = cm.ravel()
specificity = tn / (tn + fp)
print(f'Specificity: {specificity:.2f}')
```

```python
report = classification_report(y_test, y_pred, output_dict=True)
```

```python
print(report)
```

```python
precision_false = report['0']['precision']
print(f'Precision when the target is False: {precision_false:.2f}')
```

```python
total_support = report['0']['support'] + report['1']['support']
print(f'Total support: {total_support}')
```

```python
# F1 score when the target is True (Class 1)
f1_score_true = report['1']['f1-score']
print(f'F1 Score when the target is True: {f1_score_true:.2f}')

# Correctly classified records
correctly_classified = tn + tp
print(f'Correctly classified records: {correctly_classified}')
```
